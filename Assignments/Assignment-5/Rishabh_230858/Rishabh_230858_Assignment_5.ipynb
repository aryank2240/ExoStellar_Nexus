```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('/content/hwc.csv')
data
```

```python
# df = data[['P_MASS','P_RADIUS','P_PERIOD','P_SEMI_MAJOR_AXIS','P_ECCENTRICITY','P_INCLINATION','P_TEMP_SURF','P_TEMP_EQUIL','P_GRAVITY','P_DENSITY','S_TEMPERATURE','S_LUMINOSITY','S_RADIUS','S_MASS','S_HZ_OPT_MIN','S_HZ_OPT_MAX','S_HZ_CON_MIN','S_HZ_CON_MAX','P_ESI']]

df = data[['P_RADIUS','P_PERIOD','P_SEMI_MAJOR_AXIS','P_ECCENTRICITY','P_INCLINATION','P_TEMP_EQUIL','S_TEMPERATURE','S_RADIUS','S_MASS','P_ESI']]
df
```

```python
# df.dropna(inplace=True)
# df.reset_index(drop=True,inplace=True)

'''by dropping off the null values the r2_score was 0.96 only
so we replace the NaNs by mean values'''

mean_values = df.mean()
df.fillna(mean_values,inplace=True)
```

```python
x = df.drop('P_ESI',axis=1).values
y = df['P_ESI'].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
pipeline = make_pipeline(StandardScaler(),RandomForestRegressor())
pipeline.fit(x_train,y_train)
```

```python
predictions = pipeline.predict(x_test)
print('R2 Score = ',r2_score(y_test,predictions))
print('Mean Squared Error = ',mean_squared_error(y_test,predictions))
```

```python
# features = ['P_MASS','P_RADIUS','P_PERIOD','P_SEMI_MAJOR_AXIS','P_ECCENTRICITY','P_INCLINATION','P_TEMP_SURF','P_TEMP_EQUIL','P_GRAVITY',
            # 'P_DENSITY','S_TEMPERATURE','S_LUMINOSITY','S_RADIUS','S_MASS','S_HZ_OPT_MIN','S_HZ_OPT_MAX','S_HZ_CON_MIN','S_HZ_CON_MAX']
features = ['P_RADIUS','P_PERIOD','P_SEMI_MAJOR_AXIS','P_ECCENTRICITY','P_INCLINATION','P_TEMP_EQUIL','S_TEMPERATURE','S_RADIUS','S_MASS']
regressor = pipeline.named_steps['randomforestregressor']
importances = regressor.feature_importances_
feature_importance_df = pd.DataFrame({'Feature': features, 'Importance': importances})
print(feature_importance_df)
```

#KEPLER data analysis

```python
kepler_data = pd.read_csv('/content/kepler.csv')
kepler_data
```

```python
kepler_data = kepler_data[kepler_data['koi_disposition']=='CONFIRMED']
kepler_data.dropna(inplace=True)
kepler_data.reset_index(drop=True,inplace=True)
kepler_data
```

```python
kepler = kepler_data[features]
kepler
```

```python
kepler_predictions = pipeline.predict(kepler)
kepler_predictions = pd.Series(kepler_predictions)
kepler_predictions.sort_values(ascending=False,inplace=True)
kepler_predictions
```

```python
kepler_top_10 = kepler_predictions.head(10)
kepler_top_10 = kepler_data.iloc[kepler_top_10.index]
kepler_top_10['ESI'] = kepler_predictions
kepler_top_10.reset_index(drop=True,inplace=True)
kepler_top_10
```
